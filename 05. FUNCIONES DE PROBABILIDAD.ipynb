    - Media = mu
    - D.Standar = ds

## --- Simulación de Monte Carlo (v42)

* __Pasos para generar la simulación__


1. Generamos dos números aleatorios 'x' e 'y' entre 0 y 1.
2. Calculamos var = x^2 + y^2.
 - Si (var<1): Estamos dentro del círculo
 - Si (var>1): Estamos fuera  del círculo
3. Dividimos el n° de veces que var es menor que 1 entre el n° total de intentos.
4. Con esto tenemos una aproximación de la probabilidad de caer dentro del círculo.
5. Usamos dicha aprox. para calcular el valor de pi.
6. Repetimos el experimento un número suficiente de veces.
7. calculamos el promedio de las aproximaciones para dar un un valor final de PI.

<img src="../images/calculo.de.pi.png"  width="800">

```python
n_ptos = 10000
n_exps = 200
pi_avg = 0
pi_values_list = []

for i in range(n_exps):
    favorable = 0
    x = np.random.uniform(0,1,n_ptos).tolist()
    y = np.random.uniform(0,1,n_ptos).tolist()
    for j in range(n_ptos):
        z = np.sqrt(x[j]*x[j] + y[j]*y[j])
        if z<=1:
            favorable += 1
    favorable_float = float(favorable)
    pi_value = (favorable_float/n_ptos)*4
    pi_values_list.append(pi_value)
    pi_avg += pi_value

pi = pi_avg/n_exps
print(pi)
            

```
