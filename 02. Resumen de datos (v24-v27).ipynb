## --- Tratamiento de valores faltantes (v26)

* __BORRADO DE VALORES FALTANTES__

````python
data_titanic_2 = data_titanic.dropna (axis=0,how='all')
data_titanic_3 = data_titanic.dropna (axis=0,how='any')
````

```
axis=0 : Elimina filas
axis=1 : Elimina columnas
how='all' : Elimnina la fila/columna si TODOS los valores en ella son NA's
how='any' : Elimnina la fila/columna ALGUNO de los valores en ella es NA.
```
En este caso..
* data_titatic_2 mantiene todas las filas ya que ningún registro tienen todos sus valores como NA
* data_titatic_3 está vacío, puesto que por los menos uno de los valores en cada registro tiene un valor NA

** La función NO modifica al dataset en cuestión. Si desea mantener el resultado, este se debe asignar a una variable.

````python
display(data_titanic.tail())
display(data_titanic_3.tail())
````

* __REEMPLAZO DE VALORES FALTANTES__

````python
# Llena TODOS los valores faltantes en el dataset con '0' o 'Desconocido' respectivamente.
data_titanic_4 = data_titanic.fillna(0) 
data_titanic_5 = data_titanic.fillna('Desconocido')
# el dataset origen NO se modifica. El resultado es asignado a una nueva variable.

display(data_titanic.tail())
display(data_titanic_4.tail())
````

````python
# Llena SÓLO los valores faltantes en las columnas 'body' y 'home.dest' del dataset con '0' o 'Unknow' respectivamente.

data_titanic_6 = data_titanic['body'].fillna(99999) 
data_titanic_7 = data_titanic['home.dest'].fillna('Unknow')
# data_titanic_6 y data_titanic_7 guardan los resultados de las COLUMNAS sobre las que se aplicó la función.
# El dataset origen NO se modifica. El resultado es asignado a una nueva variable.

data_titanic['body'].fillna(99999)
# NO afecta al data set original

display(data_titanic.tail())
display(data_titanic_7.tail())
````

````python
data_titanic_8 = data_titanic
data_titanic_8['body'] = data_titanic['body'].fillna(99999) 
data_titanic_8['home.dest'] = data_titanic['home.dest'].fillna('Unknow')
# Llena los valores faltantes de dataset en las COLUMNA 'body' y 'home.dest' con '99999' o 'Unknow' respectivamente.
# El resultado se está guardando sobre una variable creada previamente por asignación simple "B = A".
````

````python
# Al modificarse la variable B, la variable A también se verá afectada. 
display(data_titanic_8.tail())
display(data_titanic.tail())
````

````python
# Para evitar que el dataset original se vea afectado, se debe usar la función COPY
data_titanic = pd.read_csv(fullpath)
data_titanic_9 = data_titanic.copy() # COPY
data_titanic_9['body'] = data_titanic['body'].fillna(99999) 
data_titanic_9['home.dest'] = data_titanic['home.dest'].fillna('Unknow')

display(data_titanic.tail())
display(data_titanic_9.tail())
````

````python
# REEMPLAZO DE VALORES FALTANTES POR LA MEDIA DE LOS VALORES PRESENTES
data_titanic_10 = data_titanic.copy()
pd.isnull(data_titanic_10['age']).values.ravel().sum() # Cantidad de nulos
````

````python
# Completamos los valores nulos en AGE con el promedio de las edades conocidas.
data_titanic_10['age'] = data_titanic['age'].fillna(data_titanic['age'].mean())
display(data_titanic_10.tail())
````

````python
# LLENADO CON LOS MÉTODOS "FFILL" Y "BACKFILL"

data_titanic_11 = data_titanic.copy

# - Ubicamos un valor na
print('--- Ubicamos un valor na:')
display(data_titanic['age'][1291])

# - Ubicamos el valor dentro de la columna del dataset original
print('--- Ubicamos el valor dentro de la columna del dataset original:')
display(data_titanic['age'].tail(20))
# Verificamos que el primer valor hacia atrás es 28.5 y hacia adelante es 47.

# - Llenamos los valores faltantes  con el primer valor hacia ATRÁS
print('--- Llenamos los valores faltantes  con el primer valor hacia ATRÁS:')
display(data_titanic['age'].fillna(method='backfill').tail(20))

# - Llenamos los valores faltantes  con el primer valor hacia ADELANTE
print('--- Llenamos los valores faltantes  con el primer valor hacia ADELANTE:')
display(data_titanic['age'].fillna(method='ffill').tail(20))
````

````python
# Unimos el dataframe original con el dataframe que contiene las variables dummies
data_titanic_12 = pd.concat([data_titanic,dummy_sex], axis=1)
data_titanic_12.head()
````

````python
# Podemos hacer todos los pasos anteriores en una función
def createDummies (df, var_name):
    dummies = pd.get_dummies(df[var_name], prefix = var_name)
    df = df.drop(var_name, axis = 1)
    df = pd.concat([df, dummies], axis = 1)
    return df
````

````python
# Llamamos a la función y asignamos el resultado a una nueva variable
data_titanic_13 = createDummies(data_titanic, 'sex')
data_titanic_13.head()
````
